```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#sns.set_style('whitegrid')
#sns.set_palette('winter', 8)
```

```python
np.random.seed(0)

df = pd.DataFrame({
    'id': np.random.randint(low=1, high=3000, size=100),
    'cate': np.random.randint(low=1, high=50, size=100),
    'payment': np.random.randint(low=100, high=90000, size=100)
})

display(df.info(), df.head())
```

## 基本統計量

```python
df[['payment']].describe()
```

```python
df.groupby('cate')[['payment']].count().rename(columns={'payment':'payment_cnt'}).head()
```

```python
df.groupby('cate')[['payment']].describe()
```

# seaborn

## distplot()
distplot()は１つの変数データについての分布を示す

```python
#sns.set_palette("Greens", 1)
sns.distplot(df['payment'], bins=20)
```

```python
# norm_hist=True でヒストグラムを正規化(デフォルト?)
sns.distplot(df['payment'], bins=20, norm_hist=True)
```

```python
# hist=False でヒストグラムを描画しないように設定
sns.distplot(df['payment'], bins=20, hist=False)
```

```python
# カーネル密度関数を表示しない場合
sns.distplot(df['payment'], bins=20, kde=False)
```

```python
# 複数のヒストグラムを描画
ax = sns.distplot(a=df[df['cate']==1]['payment'], kde=True)
sns.distplot(a=df[df['cate']==3]['payment'], ax=ax, kde=True)
```

## jointplot()
jointplot()は２つの変数データを組み合わせて表示して比較することができる

グラフの種類としては“scatter'、“reg'、“resid'、“kde'、“hex'

相関係数とp値も表示

```python
# scatter
sns.jointplot('payment', 'cate', data=df)
```

```python
# 六角形のヒートマップ
sns.jointplot('payment', 'cate', data=df, kind='hex')
```

```python
# 等高線
sns.jointplot('payment', 'cate', data=df, kind='kde')
```

## pairplot()
pairplot()は、データ全体の対の関係を視覚化することができる

```python
# 少し処理が重い
sns.pairplot(df, hue='cate')
```

```python
sns.pairplot(df, hue='cate', diag_kind='kde')
```

## barplot()

```python
# 黒い細い棒の長さは標準偏差の大きさ?
plt.figure(figsize=(12,5))
sns.barplot(x='cate', y='payment', data=df)
```

## kdeplot()
1変数 or 2変数の分布をkdeで可視化

```python
# 1変数
sns.kdeplot(df['payment'], shade=True)
```

```python
sns.kdeplot(df['payment'], shade=True, bw=1)
sns.kdeplot(df['payment'], shade=True, bw=10000)
```

```python
sns.kdeplot(df['payment'], shade=True, cut=0.5)
```

```python
# 2変数
sns.kdeplot(df[df['cate']==1]['payment'], df[df['cate']==3]['payment'], shade=True)
```

## 箱ひげ図

```python
plt.figure(figsize=(18,5))
sns.boxplot(x="cate", y="payment", data=df)
```
