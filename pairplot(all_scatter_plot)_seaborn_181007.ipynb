# 数値型かつ欠損値なしデータ全体の散布図を描く
データに含まれるすべての数値型データの相関を見たい場合はpairplotメソッドを使う

```python
# 数値型かつ欠損値なしカラムを抽出
df_train_numetric = df_train[['y', 'soldout', 'temperature']]
```

```python
sns.pairplot(df_train_numetric)
```

```python
df_train2 = df_train[['y', 'week','soldout', 'temperature']]
```

```python
# hueをしていることでカテゴリー変数のカラムの値ごとに色分けをすることが可能
sns.pairplot(df_train2, hue='week')
```
