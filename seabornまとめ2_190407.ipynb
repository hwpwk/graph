```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
#sns.set_palette('winter', 8)
```

```python
np.random.seed(0)

df = pd.DataFrame({
    'flag': np.random.randint(low=0, high=2, size=200),
    'cate': np.random.randint(low=1, high=20, size=200),
    'payment': np.random.randint(low=100, high=90000, size=200)
})

df.head()
```

```python
df[['payment']].describe()
```

```python
df.groupby('cate')[['payment']].count().rename(columns={'payment':'payment_cnt'}).head()
```

```python
df.groupby(['cate', 'flag'])[['payment']].describe()
```

## seaborn

## factorplot()

```python
plt.figure(figsize=(15,15))
#sns.factorplot(x='class', y='age', data=df2, hue='alive', kind='box', col='sex', row='alone')
sns.factorplot(x='cate', y='payment', data=df, hue='flag', kind='box')
```

## swarmplot()

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x=df['cate'], y=df['payment'])
```

## violinlot()
swarmplot()boxplot()

```python
plt.figure(figsize=(18,6))
sns.violinplot(x=df['cate'], y=df['payment'])
```

```python
# flag1
plt.figure(figsize=(18,6))
sns.violinplot(x=df['cate'], y=df['payment'], hue=df['flag'])
```

```python
# flag2
plt.figure(figsize=(18,6))
sns.violinplot(x=df['cate'], y=df['payment'], hue=df['flag'], split=True)
```

## lvplot()
boxplotboxplot()

```python
plt.figure(figsize=(18,6))
sns.lvplot(x=df['cate'], y=df['payment'])
```

## barplot()


```python
plt.figure(figsize=(18,6))
sns.barplot(x=df['cate'], y=df['payment'], hue=df['flag'])
```
